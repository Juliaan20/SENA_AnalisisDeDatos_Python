### Paso 2: planteamiento de pregunta objetivo.


 **¿Que variables o elementos relacionados con los precios de inmuebles en las areas del terreno construidas en la localidad?**

1. Tamaño del terreno: El área del terreno construido puede estar correlacionada con el tamaño total del terreno. A medida que aumenta el tamaño del terreno, es posible que también se incremente el precio del inmueble.

2. Ubicación geográfica: La ubicación de la localidad puede influir en los precios de los inmuebles. Por ejemplo, una localidad cercana a áreas urbanas desarrolladas o con buena infraestructura puede tener precios más altos.

3. Demanda y oferta: La relación entre la oferta y la demanda de inmuebles en una localidad puede afectar los precios. Si la demanda supera la oferta, es probable que los precios sean más altos, y viceversa.

4. Amenidades y servicios cercanos: La disponibilidad de amenidades y servicios cercanos, como parques, escuelas, centros comerciales, hospitales y transporte público, puede aumentar el atractivo de una localidad y, por lo tanto, influir en los precios de los inmuebles.

5. Infraestructura y desarrollo urbano: Una buena infraestructura puede aumentar la conveniencia y la calidad de vida, lo que puede reflejarse en los precios.

6. Características del inmueble: Otros factores relacionados con el inmueble en sí, como la edad, el estado de conservación, la calidad de la construcción, las comodidades y las características arquitectónicas, pueden influir en los precios.

```python
import seaborn as sns
import matplotlib.pyplot as plt

def barrasM(df, colx, coly, cruce):
    plt.figure(figsize=(12, 8))
    sns.set(style="whitegrid")
    sns.barplot(x=colx, y=coly, hue=cruce, data=df, palette="muted")
    plt.title("Distribución de precios por tipo de inmueble y ciudad", fontsize=16)
    plt.xlabel("Tipo de Inmueble", fontsize=14)
    plt.ylabel("Precio", fontsize=14)
    plt.legend(title="Ciudad", loc="upper right")
    plt.xticks(rotation=45)
    plt.tight_layout()

# Llamada a la función
barrasM(df, 'Tipo de Inmueble', 'Precio', 'Ciudad')
plt.show()

```

**Diagrama de torta**

```python
import matplotlib.pyplot as plt

def diagrama_torta(df, col):
    plt.figure(figsize=(8, 8))
    plt.pie(df[col].value_counts(), labels=df[col].unique(), autopct='%1.1f%%', startangle=90)
    plt.axis('equal')
    plt.title("Distribución de " + col)
    plt.show()

# Llamada a la función
diagrama_torta(df, 'Tipo de Inmueble')
```

```python
import matplotlib.pyplot as plt

def histograma(df, col):
    plt.figure(figsize=(10, 6))
    plt.hist(df[col], bins=10, edgecolor='black')
    plt.title("Histograma de " + col)
    plt.xlabel(col)
    plt.ylabel("Frecuencia")
    plt.show()

# Llamada a la función
histograma(df, 'Precio')
```
